### 1.1. Write an algorithm to search climate data for the records on 15th December 2017.
For this task, we need use the parallel search to find records on `15th December 2017`. By checking `ClimateData` (read from ClimateData.csv and stored in memory as list) using `checkSorted` function, we can know `ClimateData` is sorted on `Date` and does have duplicated records, which provides us a favorable premise to use Range Partition. The main benefit of this partitioning method is that records on a specific range can be directed to certain processors, and since we need to find the exact-matching record(s), there is only one processor will be activated when searching. Apart from this, partitioning non-duplicated and sorted data means load balance, even though Range Partition is used.  
Based on the characters of `ClimateData`, we can apply the Binary Search to get the exactly-match record on the date `15th December 2017`. The advantage of Binary Search is that it can provide constant complexity O (log n), while Linear Search may have O (n) in the worst case.
To sum up, we believe `Range Partition + Binary Search` is an ideal solution to this task.

In this assignment, we assume there are 4 processors in the computer (which is a common system hardware environment in recent days), we apply the parallel algorithm with share memory architecture, we should have 3 processors to use pool to calculate local results, and one `master processor` to control system and global aggregation. The reason to choose 4 processors is that quad-core CPU can be satisfied in most of nowadays computer, and the speed-up for 3 can reach a relatively ideal state in parallel computing (according to [the Amdahl's law](https://en.wikipedia.org/wiki/Amdahl%27s_law))

```python
def checkSorted(dataset, column):
    result = False
    
    if dataset == ClimateData:
        col = ClimateData_column
    else:
        col = FireData_column
    # find the index of the attribute used to be partitioned and searched
    for i,name in enumerate(col):
        if column == name.strip():
            index = i
            
    for x in range(1,len(dataset)):
        if dataset[x-1][index] < dataset[x][index]:
            result = True
        else:
            result = False
            break
    return result
```

```python
def range_partition_1(data,range_indices,index):
    results = []   
    # Calculate the number of bins
    n_bin = len(range_indices) #n_bin=2
    
    # For each bin, perform the following
    for i in range(n_bin): 
        # Find elements to be belonging to each range
        s = [x for x in data if time.strptime(x[index], '%Y-%m-%d') < time.strptime(range_indices[i], '%Y-%m-%d')]
        # find the data which have date less than range date
        results.append(s)
        # Find the last element in the previous partition
        last_element = s[len(s)-1]
        # Find the index of of the last element
        last = data.index(last_element)
        # Remove the partitioned list from the dataset
        data = data[int(last)+1:]
    # Append the last remaining data list
    #results.append([x for x in data if time.strptime(x[index], '%Y-%m-%d') >= time.strptime(range_indices[i], '%Y-%m-%d')])
    results.append(data)
    return results
```

```python
def binary_search(data,key,index):

    #return none if there is no match
    matched_record = None
    
    lower = 0
    middle = 0
    upper = len(data)-1
    
    while (lower <= upper):
        # calculate middle: the half of lower and upper
        middle = int((lower + upper)/2) 
        
        if (key == data[middle][index]): # if query equal data in particular index 
            # if we find the matched one
            # position = middle
            matched_record = data[middle]
            break   # stop searching
        elif key > data[middle][index]:
            # reduce to the top half of the list
            lower = middle + 1
        else:
            # reduce to the bottom half of the list
            upper = middle - 1
    
    return matched_record
```

```python
def parallel_search_1_1(data, query, n_processor,column):
    
    results = []
    
    # find the index of the attribute used to be partitioned and searched
    for i,name in enumerate(ClimateData_column):
        if column == name.strip():
            index = i 
    
    # create pool for multiple processors
    pool = mp.Pool(processes=n_processor) 
    
    # range partitioning
    Dateset = range_partition_1(data,['2017-4-1','2017-8-1'],index) 
    for d in Dateset: 
        for record in d:
            # Find the range that may contain the every
            if time.strptime(query, '%Y-%m-%d') == time.strptime(record[index], '%Y-%m-%d'):
                # apply binary search using pool
                result = pool.apply(binary_search, [d, query, index]) 
                results.append(result)
                break

    print(results)
```

Oher partitioning method, such as Round Robin and Hash Partitioning, can partition the data into relatively equal-size partitions, which means every processor will have relatively balanced workload. For the range search purpose, we choose hash partitioning method, because the benefit of the hash partitioning is that find out which processor we should activate when we search in a certain range if the range has also been hashed in the same way.
In terms of search method, we should use Linear Search method, because there may be multiple records with same surface temperature() and the `FireData` is not sorted based on surface temperature. 

```python
def s_hash(x, n):
    # simple hash using reminder, n can be number of processors
    result = x % n 
    return result
```

```python
def hash_partition_1_2(data,n,index):
    # Use the dictionary to store hash partioning 
    dic = {} 
    # For each data record, perform the following
    for x in data: 
        # Get the hash key of the input
        h = s_hash(int(x[index]),n) 
        # If the key exists
        if (h in dic.keys()): 
            # set a list to store the value of s_key
            s = dic[h] # s:[[]]
            # add the new entry to the end of the list
            s.append(x) 
            # update the value of s_key
            dic[h] = s 
        # If the key does not exist
        else: 
            # Create an empty value set
            s = [] 
            # Add the new entry to the list
            s.append(x)
            # create a new dictionary to store the key-value pair
            dic[h] = s 
    return dic
```

## Task 2: Parallel Join
### 1. Write an algorithm to find  surface temperature (C), air temperature (C), relative humidity  and  maximum wind speed.
In this task, we need to join two data tables based on the date. Firstly, consider about the two tables, the data in `ClimateData` have been sorted by date, while the data in the `FireData` has not yet sorted. In this situation, Sorted-Merge parallel Join may not quite suitable, because if we sorted `FireData` at the beginning, the complexity of the algorithm will increase O(n log n) . For the Nested-Loop Join, the complexity even more higher, O(n^2). As a result, we choose Hash-Based Join, which has only O(n) complexity for each partition. In order to hash the date, we can just simply transfer `Date` into int, (for example, from `2017-04-01` to `20170401`), then run the hash function (reminder calculation is not complex for modern computer).
In addition, we choose the Hash-Partitioning (which is one of Disjoint partitioning) to avoid broadcasting one table to all processor. Using same hash function to hash partition two tables on `date` can ensure that the records from these two tables in same date range will be locate in same processor. Then, apply Hash-Based Join logically.

```python
def hash_partition_2_1(data,index,n):
    
    result = []
    
    for i in range(n):
        result.append([])
    
    # For each data record, perform the following
    for d in data: 
        # The format of date is 'YYYY-MM-DD', remove '-' from each date and convert it to int(from str)
        date = int(''.join(d[index].split('-')))
        # Get the hash key of the input
        hash_key = date % n
        # Add the value set to the key
        result[hash_key].append(d)
    
    return result #list
```

```python
def HB_join_2_1(T1, T2):
    
    result = []

    # Use dictionary to store hash keys and the records with each hash key
    dic = {} 
    
    # For each record in table T2
    for s in T2:
        # Hash the record based on join attribute value(date) using hash function 's_hash' into hash table
        # remove '-' from each date and convert it to int(from str)
        date_s = int(''.join(s[6].split('-')))
        # hash date using hash function 's_hash'
        s_key = s_hash(date_s,3)
        # if s_key exists
        if s_key in dic.keys():
            # set a list to store the value of s_key
            templist = dic[s_key]
            # add the new entry to the end of the list
            templist.append(s)
            # update the value of s_key
            dic[s_key] = templist
        # If s_key does not exist
        else:
            # Create an empty list
            templist = []
            # Add the new entry to the list
            templist.append(s)
            # create a new dictionary to store the key-value pair
            dic[s_key] = templist
            
    # For each record in table T1 (probing)
    for r in T1:
        # Hash the record based on join attribute value(date) using 's_hash'
        # remove '-' from each date and convert it to int(from str)
        date_r = int(''.join(r[1].split('-')))
        # hash date using hash function 's_hash'
        r_key = s_hash(date_r,3)
        # If an index entry is found Then
        if r_key in dic.keys():
            # Compare each record on this index entry with the record of table T1
            for item in dic[r_key]:
                if item[6] == r[1]:
                    # Select the data we neeed and put the result
                    result.append([r[1],item[7],r[2],r[3],r[5]])

    return result
```

```python
#Perform a disjoint partitioning-based parallel join algorithm.
import multiprocessing as mp

def DPBP_join_2_1_H(T1, T2, n_processor, column):
    
    # find the index of the attribute used to be partitioned and searched
    for i,name in enumerate(ClimateData_column):
        if column == name.strip():
            climate_index = i
    for i,name in enumerate(FireData_column):
        if column == name.strip():
            fire_index = i
            
    result = []
    
    # Partition T1 & T2 into sub-tables using hash_partition().
    # pass date as variable
    T1_subsets = hash_partition_2_1(T1, climate_index, n_processor)
    T2_subsets = hash_partition_2_1(T2, fire_index, n_processor)
    
    pool = mp.Pool(processes = n_processor)
    
    for i in range(len(T1_subsets)): #len(T1_subsets)=len(T2_subsets) 0,1,2
        # Apply a HB_join on each processor
        result.append(pool.apply(HB_join_2_1, [T1_subsets[i], T2_subsets[i]]))

    return result
```

## Task 3: Parallel Sort
### 1. Write an algorithm to sort fire data based on surface temperature(C) in an ascending order.
In the local sort for each partitioning, we can use serial external sorting, the reason is when the dataset is too large to fit into main memory even for partitions, we need to use this method to store temporary results to disk (we use simulate file: list as storage in this task), then add them together. The choice of buffer size is important in external sort, for our task, we have about 1000 records in each partition, according to the Textbook, when Buffer size = 9, 1000 records need only 4 passes which can be a reasonable value. With the increasing numbers of records, the buffer size should be considered carefully by two aspects: the size that can fit to main memory (influence the size of buffer), and the number of passes based on buffer size and number of records (influence the calculation efficiency).

For this task, we need to sort the `FireData` based on the surface temperature (), consider about the structure of our data, each records store in a list, and records store in a larger list. Because of the volume of the data, we need to partition it at first. For the sorting process, we should sort all the records, for this reason, Round-Robin partitioning can be used since it divides the data evenly that can ensure balance load.

After partitioning, we can perform the sorting algorithm. We apply quick sort inside the partition, since compared with bubble sorting or insertion sorting, the quick sorting is a divide and conquer algorithm which has better complexity. After sorting each partition, we use the k-way merge to merge all the sorted partitions. The benefit of using k-way merge is that if we have multiple buffers to save the subset, the smallest value in each partition can be found at the same time, which is quite efficient.


```python
#quick sort algorithm
def qsort(arr,index): 

    if len(arr) <= 1:
        return arr
    else:
        #recurcive sort the arr
        return qsort([x for x in arr[1:] if int(x[index]) < int(arr[0][index])],index) \
                + [arr[0]] \
                + qsort([x for x in arr[1:] if int(x[index]) >= int(arr[0][index])],index)
            
```

```python
#find the minium among partitionings
def find_min(records):  
    m = int(records[0])
    index = 0
    for i in range(len(records)):
        if(int(records[i]) < m):  
            index = i
            m = int(records[i])
    return index

#k-way mearge method 
def k_way_merge(record_sets,index): 
    indexes = []
    for x in record_sets:
        # initialisation with 0
        indexes.append(0) 

    # final result will be stored in this variable
    result = []  
    
    # the merging unit (i.e. # of the given buffers)
    tuple = []
    
    while(True):
        # initialise tuple
        tuple = [] 
        
        # This loop gets the current position of every buffer
        for i in range(len(record_sets)):
            if(indexes[i] >= len(record_sets[i])):
                tuple.append(sys.maxsize)
            else:
                tuple.append(record_sets[i][indexes[i]][index])  
        
        # find the smallest record 
        smallest = find_min(tuple) 
        #break if there is only sys.maxsize in the tuple
        if(tuple[smallest] == sys.maxsize): 
            break  
        result.append(record_sets[smallest][indexes[smallest]])
        indexes[smallest] +=1
   
    return result
```

```python
#perform round-robin partitioning
def rr_partition(data, n): 
    result = []
    # divide data in to n partitions
    for i in range(n): 
        result.append([])
    # calculate the number of the elements to be allocated to each bin
    n_bin = len(data)/n
    
    # for each bin, perform the following
    for index, element in enumerate(data): 
        # calculate the index of the bin that the current data point will be assigned
        index_bin = (int) (index % n)
        result[index_bin].append(element)
    return result
```

```python
#local sort for partitions
def serial_sorting(dataset,buffer_size,index): 
    # return if buffer less than 2
    if (buffer_size <= 2):
        print("Error: buffer size should be greater than 2")
        return 
    
    result = [] 
    
    # --- Sort Phase ---
    sorted_set = []
    start_pos = 0
    N = len(dataset)
    while True:
        #read the data part by part (part equals to buffer size)
        if ((N - start_pos) > buffer_size): 
            #create subset
            subset = dataset[start_pos:start_pos + buffer_size] 
            #use quick sort for each subset 
            sorted_subset = qsort(subset,index) 
            #store the results to sorted_set
            sorted_set.append(sorted_subset) 
            start_pos += buffer_size
        else:
            # read the last B-records from the input, where B is less than buffer_size
            subset = dataset[start_pos:] 
            # use quick sort for each subset 
            sorted_subset = qsort(subset,index) 
            sorted_set.append(sorted_subset)
            break
    
    # --- Merge Phase ---
    merge_buffer_size = buffer_size - 1
    dataset = sorted_set
    while True:
        merged_set = []
        N = len(dataset)
        start_pos = 0
        while True:
            # read the data part by part(part equals to buffer size)
            if ((N - start_pos) > merge_buffer_size): 
                subset = dataset[start_pos:start_pos + merge_buffer_size]
                # merge lists in subset
                merged_set.append(k_way_merge(subset,index)) 
                start_pos += merge_buffer_size
            else:
                subset = dataset[start_pos:]
                # merge lists in subset
                merged_set.append(k_way_merge(subset,index)) 
                break

        dataset = merged_set
        # if the size of merged record set is 1, then stop 
        if (len(dataset) <= 1): 
            result = merged_set
            break
    return result
```

```python
def parallel_merge_all_sorting(dataset, n_processor, buffer_size, column):
    
    # return if buffer less than 2
    if (buffer_size <= 2):
        print("Error: buffer size should be greater than 2")
        return 
    
    #find the index match the column name
    for i,name in enumerate(FireData_column):
        if column == name.strip():
            index = i 
            
    result = []
    
    #use round-robin partitioning
    subsets = rr_partition(dataset, n_processor)#use round-robin partitioning
    
    # create pool
    pool = mp.Pool(processes = n_processor)  

    # ----- Sort phase -----
    sorted_set = []
    for s in subsets:
        #append agruement as list
        sorted_set.append(*pool.apply(serial_sorting, [s,buffer_size,index])) 
    pool.close()
    
    # ---- Final merge phase ----
    #merge all the results using k-way merge
    result = k_way_merge(sorted_set,index) 
    return result
```

## Task 4: Parallel Group-By
### 4.1. Write an algorithm to get the number of fire in each day. You are required to only display total number of fire and the date in the output.
For this task, we use the two-phase parallel group-by. In first step, we choose the Round-Robin partitioning because it can provide equal workload for each processor, and we can except similar run-time for local group-by count in this situation. The local results can be stored as a local result dictionary with count as value and group-by attribute as the key. In the global aggregation phase, we only need to use a dictionary create the key-value pair, and calculate the results easily, because the update of dictionary only has complexity of O(1).

```python
def local_groupby_4_1(dataset,index):

    dic = {}
    for record in dataset:
        # group by date (the index can be passes from the variable inserted by user)
        key = record[index] 
        if key not in dic:
            dic[key] = 0
        #the operator is count, so if found, add 1 to the key
        dic[key] += 1 
    return dic
```

```python
def parallel_merge_all_groupby_4_1(dataset, column, n_processor):

    # find the index of the attribute used to be partitioned and searched
    for i,name in enumerate(FireData_column):
        if column == name.strip():
            index = i
    
    result = {}
    
    # using round-robin to partition dataset
    subsets = rr_partition(dataset, n_processor)
    
    pool = mp.Pool(processes = n_processor)

    # ----- Local aggregation step -----
    local_result = []
    for s in subsets:
        # call the local aggregation method
        local_result.append(pool.apply(local_groupby_4_1, [s,index]))
    pool.close()

    # ---- Global aggregation step ----
    # The global operator is sum.
    for r in local_result: #local_result:[{},{},{}]
        for key, val in r.items():
            if key not in result:
                result[key] = 0
            result[key] += val    
    
    return result
```

### 2. Write an algorithm to find the average surface temperature(C)  for each day. You are required to only display  average surface temperature(C) and the date in the output.
Based on the same concept of Q4.1, we choose the Round-Robin partitioning to provide equal workload for processors. Using the pool to calculate local results and merge them together by using a global dictionary. After getting the sum of the surface temperature (), and the number of fire occurring in the same day, we can get the average surface temperature() of that day (the sum of the surface temperature/ the number of fire in that day).

```python
def local_groupby_4_2(dataset,group_column,data_column):

    dict = {}
    for record in dataset:
        key = record[group_column] #date
        val = record[data_column] #ST
        if key not in dict:
            dict[key] = [0,0] #initial
        dict[key][0] += int(val)
        dict[key][1] += 1
    return dict
```

```python
def parallel_merge_all_groupby_4_2(dataset,group_column,data_column,n_processor):

    for i,name in enumerate(FireData_column):
        if group_column == name.strip():
            group_index = i
        if data_column == name.strip():
            data_index = i
    
    result = {}
 
    subsets = rr_partition(dataset, n_processor)
    
    pool = mp.Pool(processes = n_processor)

    # ----- Local aggregation step -----
    local_result = []
    for s in subsets:
        # call the local aggregation method
        local_result.append(pool.apply(local_groupby_4_2, [s,group_index,data_index])) #dic{}
    pool.close()

    # ---- Global aggregation step ----
    # The global operator is avg.
    for r in local_result:
        for key, val in r.items():
            if key not in result:
                result[key] = [0,0] #key date
            result[key][0] += val[0]
            result[key][1] += val[1]
    for key, val in result.items():
        result[key] = result[key][0]/result[key][1]
    
    return result
```

## Task 5: Parallel Group-By Join (High Distinction Only)
### 1. Write an algorithm to find the average surface temperature(C) for each weather station. You are required to only display average surface temperature(C) and the station in the output.
For this task, it can be break into two sub-processes. The first one is parallel join based on the `date` in two tables, and the second one is parallel group-by. We need to partition data in the parallel join stage and use the joined result as the input of the parallel group-by. For the reason mentioned before in Q2.1, we use the hash-partitioning based on the date in each record. After partitioning, we used the hash-based parallel join which is effective in terms of complexity. In the parallel group-by stage, we get the joined result which already have been partitioned and based on the same concept of Q4.2., we can calculate the average surface temperature() for each weather station.

```python
#hash-based join
def HB_join_5(T1, T2): 
    
    result = []
    # use a dictionary to store temp results
    dic = {} 

    # For each record in table T2
    for s in T2:
        # Hash the record based on join attribute value(data) using hash function s_hash into hash table
        date_s = int(''.join(s[6].split('-')))
        # hash date using hash function 's_hash'
        s_key = s_hash(date_s,3)
        # if s_key exists
        if s_key in dic.keys():
            # set a list to store the value of s_key
            templist = dic[s_key]
            # add the new entry to the end of the list
            templist.append(s)
            # update the value of s_key
            dic[s_key] = templist
            # if s_key does not exist
        else:
            # Create an empty list
            templist = []
            # Add the new entry to the list
            templist.append(s)
            # create a new dictionary to store the key-value pair
            dic[s_key] = templist
            
    # For each record in table T1 (probing)
    for r in T1:
        # Hash the record based on join attribute value(date) using s_hash
        date_r = int(''.join(r[1].split('-')))
        r_key = s_hash(date_r,3)
        # If an index entry is found Then
        if r_key in dic.keys():
            # Compare each record on this index entry with the record of table T1
            for item in dic[r_key]:
                if item[6] == r[1]:
                    # Select the data we neeed and put the result
                    result.append([r[1],r[0],item[7]])
    return result
```

```python
def local_groupby_sum_5(dataset,group_index,data_index):
    
    # use dictionary to store the sum of surface temperature and the count of entries
    dic = {} 
    
    #search every records in dataset
    for record in dataset: 
        key = str(record[group_index]) #date
        val = record[data_index]
        # if key not in dictionary, create new entry
        if key not in dic:
            dic[key] = [0,0] 
        # if found, add value to the key
        dic[key][0] += int(val)
        # count the number of entries to the key
        dic[key][1] += 1 
    return dic
```

```python
def parallel_merge_all_groupby_5(dataset,group,data,n_processor):

    column_name = ['Date','Station','Surface Temperature (Celcius)']
    for i,item in enumerate(column_name):
        # get the index of the attribute on which we group by
        if group == item:
            group_index = i 
        # get the index of attribute(s) aggregated
        if data == item:
            data_index = i 
    result = {}

    # create pool
    pool = mp.Pool(processes = n_processor)

    # ----- Local aggregation step -----
    local_result = []
    for s in dataset:
        # get local group-by results in parallel
        local_result.append(pool.apply(local_groupby_sum_5, [s,group_index,data_index]))
    pool.close()

    # ---- Global aggregation step ----
    for r in local_result:
        for key, val in r.items():
            if key not in result: 
                result[key] = [0,0]
            # calculate the sum of the temperature
            result[key][0] += val[0] 
            # count the number of the entries
            result[key][1] += val[1] 
    for key, val in result.items():
        #caculate the average temperature of each station
         result[key] = result[key][0]/result[key][1]  

    return result
```

```python
def parallel_join_groupby_5(T1, T2, n_processor, join_base, group_by, avg_data):
    
    result = []
    
    for i,name in enumerate(ClimateData_column):
        if join_base == name.strip():
            climate_index = i
    for i,name in enumerate(FireData_column):
        if join_base == name.strip():
            fire_index = i

    # hash partitioning based on date
    T1_subsets = hash_partition_2_1(T1, climate_index, 3)
    T2_subsets = hash_partition_2_1(T2, fire_index, 3) 
    
    pool = mp.Pool(processes = n_processor)#create pool
    
    for i in range(len(T1_subsets)):
        result.append(pool.apply(HB_join_5, [T1_subsets[i], T2_subsets[i]]))
    
    return parallel_merge_all_groupby_5(result,group_by,avg_data,n_processor)
```
